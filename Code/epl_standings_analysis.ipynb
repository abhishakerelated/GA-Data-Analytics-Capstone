#### During Pandemic 

```python
during_covid = epl_standings[(epl_standings['Season'] >= 2016) & (epl_standings['Season'] < 2020)]
during_covid
```

#### Post Pandemic

```python
post_covid = epl_standings[epl_standings['Season'] >= 2020]
post_covid
```

#### Comparison

```python
during_covid_pts = during_covid.groupby('Club')['Pts'].sum().reset_index()
post_covid_pts = post_covid.groupby('Club')['Pts'].sum().reset_index()
```

```python
comparison_df = pd.merge(during_covid_pts, post_covid_pts, on = 'Club', suffixes = (' 2016 - 2019', ' 2020 - 2023'), how = 'outer')
comparison_df
```

#### Dropping NaN

```python
drop_comparison_df = comparison_df.dropna(subset=['Pts 2016 - 2019', 'Pts 2020 - 2023'])
drop_comparison_df
```

#### Rate of Change in Points (sorted)

```python
drop_comparison_df['Rate_Of_Change'] = ((drop_comparison_df['Pts 2020 - 2023'] - drop_comparison_df['Pts 2016 - 2019']) / drop_comparison_df['Pts 2016 - 2019']) * 100
drop_comparison_df['Rate_Of_Change'] = drop_comparison_df['Rate_Of_Change'].round(2)
sorted_drop_comparison_df = drop_comparison_df.sort_values(by='Rate_Of_Change', ascending=False)
sorted_drop_comparison_df
```

#### Pts 2016 - 2019/ Pts 2020 - 2023/ Rate of Change

```python
top_five_clubs_post_covid = sorted_drop_comparison_df

fig, ax = plt.subplots(figsize=(18, 15))

bar_width = 0.3

bar_positions_2017_2019 = range(len(top_five_clubs_post_covid))
bar_positions_2020_2023 = [pos + bar_width for pos in bar_positions_2017_2019]
bar_positions_rate_of_change = [pos + 2 * bar_width for pos in bar_positions_2017_2019]

ax.bar(bar_positions_2017_2019, top_five_clubs_post_covid['Pts 2016 - 2019'], width=bar_width, label='Pts 2016 - 2019', color='blue')

ax.bar(bar_positions_2020_2023, top_five_clubs_post_covid['Pts 2020 - 2023'], width=bar_width, label='Pts 2020 - 2023', color='orange')

ax.bar(bar_positions_rate_of_change, top_five_clubs_post_covid['Rate_Of_Change'], width=bar_width, label='Rate of Change', color='skyblue')

ax.set_xticks([pos + bar_width for pos in bar_positions_2017_2019])
ax.set_xticklabels(top_five_clubs_post_covid['Club'], rotation=90)

ax.set_ylabel('Points/ Rate of Change in %', fontsize = 15)

ax.legend(loc='center', bbox_to_anchor=(0.5, 0.90), ncol=1, fontsize = 15)

min_value = top_five_clubs_post_covid['Rate_Of_Change'].min()
max_value = top_five_clubs_post_covid['Rate_Of_Change'].max()

ax.text(bar_positions_rate_of_change[0], min_value + 2, f'Min: {min_value}%', ha='left', va='bottom', color='black', fontsize = 14)
ax.text(bar_positions_rate_of_change[-1], max_value - 5, f'Max: {max_value}%', ha='right', va='bottom', color='black', fontsize = 14)

plt.savefig('bar_chart_combined.png')

plt.show()
```

#### Top Five Clubs Post Covid

#### Points Per Game (PPG)

```python
post_covid['PPG'] = round(post_covid['Pts'] / post_covid['MP'], 2)
(post_covid[['Club', 'MP', 'Pts', 'PPG', 'Season']])
```

#### PPG for Selected Clubs

```python
selected_clubs = ['Aston Villa', 'Fulham', 'Brighton', 'West Ham', 'Newcastle']

top_clubs_post_covid = post_covid[post_covid['Club'].isin(selected_clubs)]
top_clubs_post_covid[['Position', 'Club', 'MP', 'W', 'D', 'L', 'Pts', 'PPG', 'Season']]
```

#### Top Clubs Post Covid based on Rate of Change in Points per Game

```python
fig, ax = plt.subplots(figsize = (12, 8))

for club in selected_clubs: 
    club_data = top_clubs_post_covid[top_clubs_post_covid['Club'] == club]
    ax.plot(club_data['Season'], club_data['PPG'], label = f'{club} - PPG')
    
ax.set_xlabel('Season')
ax.set_ylabel('PPG')
ax.legend()
plt.title('Top Clubs Post Covid based on Rate of Change in Points')

plt.show()
```

```python
import seaborn as sns

fig, ax = plt.subplots(figsize=(12, 8))

for club in selected_clubs:
    club_data = top_clubs_post_covid[top_clubs_post_covid['Club'] == club]
    sns.lineplot(x='Season', y='PPG', data=club_data, label=f'{club} - PPG', ax=ax)

ax.set_xlabel('Season')
ax.set_ylabel('PPG')
ax.legend()
plt.title('Top Clubs Post Covid based on Rate of Change in Points')

plt.show()
```
