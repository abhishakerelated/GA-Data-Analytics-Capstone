```python
seasons = ['2019', '2020', '2021', '2022']

dfs = []

for season in seasons:
    params = {'season': season}
    
    epl_request = requests.get(epl_url, headers=headers, params=params)
    
    if epl_request.status_code == 200: 
        epl_data = epl_request.json()
        
        standings_info = epl_data['standings'][0]['table']
        
        positions = []
        team_names = []
        played_games = []
        forms = []
        wins = []
        draws = []
        losses = []
        points = []
        goals_for = []
        goals_against = []
        goal_difference = []
        
        for team in standings_info:
            positions.append(team['position'])
            team_names.append(team['team']['name'])
            played_games.append(team['playedGames'])
            forms.append(team['form'])
            wins.append(team['won'])
            draws.append(team['draw'])
            losses.append(team['lost'])
            points.append(team['points'])
            goals_for.append(team['goalsFor'])
            goals_against.append(team['goalsAgainst'])
            goal_difference.append(team['goalDifference'])
            
        df = pd.DataFrame({
            'Position': positions,
            'Club': team_names,
            'MP': played_games,
            'W': wins,
            'D': draws,
            'L': losses,
            'GF': goals_for,
            'GA': goals_against,
            'GD': goal_difference,
            'Pts': points,
            'Last 5': forms,
            'Season': [season] * len(positions) 
        })
        
        dfs.append(df)
        
    else: 
        print(f'Error for season {season}: {epl_request.status_code}')
        
result_df = pd.concat(dfs, ignore_index = True)

result_df

```
