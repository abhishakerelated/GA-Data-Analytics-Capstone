#### English Premier League Standings (2000 - 2023)

```python
standings_2000_to_2023 = epl_standings[(epl_standings['Season_End_Year'] >= 2000) & (epl_standings['Season_End_Year'] <= 2023)]
standings_2000_to_2023
```

#### Big Six Teams

```python
big_six = {'Arsenal', 'Chelsea', 'Liverpool', 'Manchester City', 'Manchester Utd', 'Tottenham' }

big_six_standings = standings_2000_to_2023[standings_2000_to_2023['Team'].isin(big_six)]
big_six_standings
```

#### Big Six Teams Points

```python
big_six_pts = big_six_standings[big_six_standings['Team'].isin(big_six)].groupby('Team')['Pts'].sum()
big_six_pts
```

#### All Teams with Matches Played and Points

```python
all_team_pts = standings_2000_to_2023.groupby('Team').agg({'MP': 'sum', 'Pts': 'sum'}).reset_index()
sorted_team_points = all_team_pts.sort_values(by='Pts', ascending=False)
sorted_team_points
```

#### Relegated Teams from 2000 - 2023

```python
relegated_teams = standings_2000_to_2023[standings_2000_to_2023['Notes'].str.contains('Relegated', na=False)]
relegated_teams
```

#### Number of Times a Team Relegated 

```python
relegated_count = relegated_teams.groupby('Team').size().reset_index(name = 'Times_Relegated')
relegated_count
```

#### Teams Relegated by Count

```python
sorted_relegated_count = relegated_count.sort_values(by= 'Times_Relegated', ascending=False)
sorted_relegated_count
```

#### Team Ranked 1 from 2000 - 2023

```python
top_ranked_teams = standings_2000_to_2023[standings_2000_to_2023['Rk'] == 1]
print(top_ranked_teams[['Season_End_Year', 'Team', 'Rk']])
```

#### Number of Titles by Team (2000 - 2023)

```python
import matplotlib.pyplot as plt

top_ranked_teams = standings_2000_to_2023[standings_2000_to_2023['Rk'] == 1]
team_rank_count = top_ranked_teams['Team'].value_counts()

plt.figure(figsize = (10, 6))
team_rank_count.plot(kind = 'bar', color = 'skyblue')
plt.bar(top_ranked_teams['Team'], top_ranked_teams['Rk'])
plt.xlabel('Team')
plt.ylabel('Count')
plt.title('Number of Times Each Team Ranked as 1')
plt.xticks(rotation = 0, ha = 'center')

plt.show()
```

#### Top Relegated Teams (2000 - 2023)

```python
relegated_teams = standings_2000_to_2023[standings_2000_to_2023['Notes'].str.contains('Relegated', na=False, case=False)]

relegated_count = relegated_teams['Team'].value_counts()

plt.figure(figsize = (12, 6))
relegated_count.plot(kind = 'bar', color = 'coral')
plt.xlabel('Relegated Team')
plt.ylabel('Count')
plt.title('Number of Times Each Team Relegated')
plt.xticks(rotation = 90, ha= 'center')

plt.show()
```

#### Pre-Covid v. Post-Covid 

```python
pre_covid = standings_2000_to_2023[(standings_2000_to_2023['Season_End_Year'] >= 2018) & (standings_2000_to_2023['Season_End_Year'] <= 2020)]
post_covid = standings_2000_to_2023[standings_2000_to_2023['Season_End_Year'] >= 2021]

pre_covid
```

```python
post_covid
```

#### Sorted Pre-Covid v. Post-Covid

```python
pre_covid_pts = pre_covid.groupby('Team')['Pts'].sum().reset_index()
post_covid_pts = post_covid.groupby('Team')['Pts'].sum().reset_index()
```

```python
comparison_df = pd.merge(pre_covid_pts, post_covid_pts, on = 'Team', suffixes = ('_Pre-Covid 2018-2020', '_Post-Covid 2021-2023'), how ='outer')
comparison_df
```

#### Dropping NaN

```python
drop_comparison_df = comparison_df.dropna(subset=['Pts_Pre-Covid 2018-2020', 'Pts_Post-Covid 2021-2023'])
drop_comparison_df
```

#### Rate of Change in Points Pre-Covid v. Post-Covid

```python
drop_comparison_df['Rate_Of_Change'] = ((drop_comparison_df['Pts_Post-Covid 2021-2023'] - drop_comparison_df['Pts_Pre-Covid 2018-2020'])/ drop_comparison_df['Pts_Pre-Covid 2018-2020']) * 100
drop_comparison_df['Rate_Of_Change'] = drop_comparison_df['Rate_Of_Change'].round(2)
drop_comparison_df
```

#### Sorted Rate of Change in Points 

```python
sorted_drop_comparison_df = drop_comparison_df.sort_values(by='Rate_Of_Change', ascending=False)
sorted_drop_comparison_df
```

```python
top_five_teams = sorted_drop_comparison_df.head(5)

plt.figure(figsize = (10, 6))
plt.bar(top_five_teams['Team'], top_five_teams['Rate_Of_Change'], color='skyblue')
plt.xlabel('Team')
plt.ylabel('Rate of Change %')
plt.title('Top 5 Teams with the Highest Rate of Change in Points')
plt.xticks(rotation = 0, ha = 'center')
plt.tight_layout()

plt.show()
```
