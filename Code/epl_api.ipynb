```python
import requests
import pandas as pd

api_token = 'd04765b4d21a47b6a307c67df0c3162c'
epl_url = 'http://api.football-data.org/v4/competitions/PL/standings'
headers = {'X-Auth-Token': api_token}

# List of seasons you want to retrieve data for
seasons = ['2020', '2021', '2022', '2023']

# Initialize an empty list to store DataFrames for each season
dfs = []

for season in seasons:
    # Specify the season parameter
    params = {'season': season}

    # Make the request with the specified parameters
    epl_request = requests.get(epl_url, headers=headers, params=params)

    # Check if the request was successful (status code 200)
    if epl_request.status_code == 200:
        # Parse the JSON response
        epl_data = epl_request.json()

        # Extract relevant data
        standings_info = epl_data['standings'][0]['table']

        # Initialize empty lists to store data
        positions = []
        team_names = []
        played_games = []
        forms = []
        wins = []
        draws = []
        losses = []
        points = []
        goals_for = []
        goals_against = []
        goal_difference = []

        # Extract data for each team in the standings
        for team in standings_info:
            positions.append(team['position'])
            team_names.append(team['team']['name'])
            played_games.append(team['playedGames'])
            forms.append(team['form'])
            wins.append(team['won'])
            draws.append(team['draw'])
            losses.append(team['lost'])
            points.append(team['points'])
            goals_for.append(team['goalsFor'])
            goals_against.append(team['goalsAgainst'])
            goal_difference.append(team['goalDifference'])

        # Create a DataFrame for the current season
        df = pd.DataFrame({
            'Position': positions,
            'Club': team_names,
            'MP': played_games,
            'W': wins,
            'D': draws,
            'L': losses,
            'GF': goals_for,
            'GA': goals_against,
            'GD': goal_difference,
            'Pts': points,
            'Last 5': forms,
            'Season': [season] * len(positions)  # Add a 'Season' column with the current season
        })

        # Append the DataFrame to the list
        dfs.append(df)
    else:
        # Print an error message if the request was not successful
        print(f"Error for season {season}: {epl_request.status_code}")

# Concatenate DataFrames for all seasons
result_df = pd.concat(dfs, ignore_index=True)

# Print or use the concatenated DataFrame as needed
result_df
```
